```python
#HAVE the numpy array and sort binary search 5900 [212,56,2,4,5900,1,0]
import numpy as np

arr = np.array([212, 56, 2, 4, 5900, 1, 0])
target = 5900
n = len(arr)

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print("Sorted Array:", arr)

left = 0
right = n - 1
found = False
position = -1

while left <= right:
    mid = (left + right) // 2
    if arr[mid] == target:
        found = True
        position = mid
        break
    elif arr[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

if found:
    print(f"Target {target} found at index: {position}")
else:
    print(f"Target {target} not found")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Input: Body Temperature in Celsius
# 2. Output: Fever Severity Index (0 = Normal, 100 = High Fever)
body_temp = np.array([36.5, 37.0, 37.5, 38.5, 39.5, 40.0]).reshape(-1, 1)
fever_index = np.array([0, 5, 20, 50, 85, 100])

# Initialize and Train the Model
model = LinearRegression()
model.fit(body_temp, fever_index)

# Predict severity based on the temperatures
predictions = model.predict(body_temp)

# Visualization
plt.scatter(body_temp, fever_index, color='teal', label='Observed Cases')
plt.plot(body_temp, predictions, color='orange', linewidth=2, label='Fever Trend')

plt.xlabel('Body Temperature (°C)')
plt.ylabel('Fever Severity Index')
plt.title("Body Fever Detection & Trend Analysis")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#Task- Create employee dataset perform all the comands what we studied
import pandas as pd

# -------------------------
# Create Employee Dataset
# -------------------------
data = {
    'emp_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['John', 'Emma', 'Raj', 'Sophia', 'Mike', 'Anita', 'David'],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'salary': [60000, 55000, 70000, 65000, 52000, 72000, 68000],
    'age': [28, 32, 35, 30, 29, 40, 38],
    'experience': [3, 5, 8, 6, 4, 10, 9]
}

df = pd.DataFrame(data)

# -------------------------
# Basic Operations
# -------------------------
print("Head:\n", df.head())
print("Tail:\n", df.tail())
print("Shape:", df.shape)
print("Info:\n")
print(df.info())
print("Describe:\n", df.describe())

# -------------------------
# Selecting Data
# -------------------------
print("Salary Column:\n", df['salary'])
print("Name & Salary:\n", df[['name', 'salary']])
print("First Row using loc:\n", df.loc[0])
print("First 3 rows using iloc:\n", df.iloc[0:3, 1:4])

# -------------------------
# Filtering
# -------------------------
print("Salary > 60000:\n", df[df['salary'] > 60000])
print("IT Department:\n", df[df['department'] == 'IT'])
print("Male & Salary > 60000:\n", df[(df['salary'] > 60000) & (df['gender'] == 'Male')])

# -------------------------
# Add / Modify Columns
# -------------------------
df['bonus'] = df['salary'] * 0.10
df['salary'] = df['salary'] * 1.05
print("After Adding Bonus & Increasing Salary:\n", df)

# -------------------------
# Drop Column
# -------------------------
df.drop('bonus', axis=1, inplace=True)

# -------------------------
# Rename Column
# -------------------------
df.rename(columns={'emp_id': 'employee_id'}, inplace=True)

# -------------------------
# GroupBy
# -------------------------
print("Average Salary by Department:\n", df.groupby('department')['salary'].mean())
print("Total Salary by Gender:\n", df.groupby('gender')['salary'].sum())
print("Multiple Aggregations:\n", df.groupby('department')['salary'].agg(['mean','max','min']))

# -------------------------
# Sorting
# -------------------------
print("Sorted by Salary Desc:\n", df.sort_values('salary', ascending=False))

# -------------------------
# Aggregations
# -------------------------
print("Total Salary:", df['salary'].sum())
print("Average Salary:", df['salary'].mean())
print("Max Salary:", df['salary'].max())
print("Min Salary:", df['salary'].min())
print("Count Salary:", df['salary'].count())

# -------------------------
# Value Counts & Unique
# -------------------------
print("Department Count:\n", df['department'].value_counts())
print("Unique Departments:", df['department'].unique())
print("Number of Departments:", df['department'].nunique())

# -------------------------
# Apply Function
# -------------------------
def salary_level(x):
    return 'High' if x > 65000 else 'Low'

df['salary_level'] = df['salary'].apply(salary_level)
print("After Salary Level:\n", df)

# -------------------------
# Missing Values
# -------------------------
print("Missing Values:\n", df.isnull().sum())

# -------------------------
# Save CSV
# -------------------------
df.to_csv('employee_data.csv', index=False)

print("All operations completed successfully!")
```

```python
# Import libraries
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
diabetes = load_diabetes()

# Create DataFrame
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df["target"] = diabetes.target

# Basic Operations
print("First 5 rows:\n", df.head())
print("\nDataset Shape:", df.shape)
print("\nStatistical Summary:\n", df.describe())
print("\nMean target value:", df["target"].mean())
print("\nCorrelation with target:\n", df.corr()["target"].sort_values(ascending=False))

# Plot: BMI vs Target
plt.figure(figsize=(6,4))
plt.scatter(df["bmi"], df["target"])
plt.title("BMI vs Disease Progression")
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.show()
```
